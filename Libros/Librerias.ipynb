# 📚 Librerías de Python Utilizadas

## ¿Qué es una librería?

Una **librería** (también llamada **módulo** o **paquete**) es un conjunto de herramientas y funciones que otros programadores crearon para resolver problemas específicos. En lugar de escribir todo desde cero, podemos importar y usar estas herramientas.

### Cómo usar una librería:

```python
import numpy as np        # Importar toda la librería
from pandas import read_csv  # Importar solo una función
```

## 📖 Lectura recomendada: Biblioteca Estándar de Python

Python incluye muchas funciones y módulos ya instalados (sin necesidad de descargar nada). Se llama la **Biblioteca Estándar**.

**Link oficial:** https://docs.python.org/3/library/

### Módulos estándar útiles para empezar:

1. **[`time`](https://docs.python.org/3/library/time.html)** - Medir tiempo y crear pausas
2. **[`math`](https://docs.python.org/3/library/math.html)** - Funciones matemáticas básicas (sin=seno, cos, log, etc.)
3. **[`random`](https://docs.python.org/3/library/random.html)** - Generar números aleatorios
4. **[`json`](https://docs.python.org/3/library/json.html)** - Trabajar con archivos JSON
5. **[`statistics`](https://docs.python.org/3/library/statistics.html)** - Calcular media, mediana, desviación estándar

### Cómo usarlas:

```python
# Ejemplo 1: math
import math
print(math.sin(math.pi))  # seno de π

# Ejemplo 2: statistics
from statistics import mean
datos = [1, 2, 3, 4, 5]
print(mean(datos))  # 3.0
```

## 🎯 Consejos para aprender librerías

1. **Lee la documentación oficial** - La mejor manera de entender una librería es leer ejemplos en la doc
2. **Comienza con lo básico** - No intentes aprender todo de una vez
3. **Practica con ejemplos** - Copia código de ejemplos y modifícalo
4. **Usa Google** - Cuando no sepas cómo hacer algo, busca: "librería + tarea"
5. **Experimenta** - En Jupyter, prueba código nuevo sin miedo a los errores
